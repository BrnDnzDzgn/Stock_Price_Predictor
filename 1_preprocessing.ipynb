## Preprocessing

```python
import os
import pandas as pd
import numpy as np

DATASET_DIR = "dataset"
```

```python
#Checking the general outline of the data
summary = []

print("=" * 80)
print("STOCK DATASET OVERVIEW")
print("=" * 80)

for file in sorted(os.listdir(DATASET_DIR)):
    if not file.endswith(".csv"):
        continue

    path = os.path.join(DATASET_DIR, file)
    df = pd.read_csv(path)

    print(f"\nFILE: {file}")
    print("-" * 60)
    print("Rows:", len(df))
    print("Columns:", list(df.columns))
    print("Data types:")
    print(df.dtypes)
    print("Missing values:")
    print(df.isnull().sum())
    print("First 3 rows:")
    display(df.head(3))

    summary.append({
        "Stock": file.replace(".csv", ""),
        "Rows": len(df),
        "Columns": len(df.columns),
        "Column Names": ", ".join(df.columns)
    })

summary_df = pd.DataFrame(summary)

print("\n" + "=" * 80)
print("SUMMARY TABLE")
print("=" * 80)
display(summary_df)
```

```python
# Displaying the Missing Column

nvda = pd.read_csv("dataset/NVDA.csv")

# Find rows that contain at least one NaN
missing_rows = nvda[nvda.isnull().any(axis=1)]

print("Number of rows with missing values:", len(missing_rows))

# Display them
display(missing_rows)
```

```python
#dropping the empty row
nvda_clean = nvda.dropna()
nvda_clean.to_csv("dataset/NVDA.csv", index=False)
```

```python
# Building one combined dataset out of all csv
dfs = []

for file in os.listdir(DATASET_DIR):
    if not file.endswith(".csv"):
        continue

    stock = file.replace(".csv", "")
    df = pd.read_csv(os.path.join(DATASET_DIR, file))
    df["Stock"] = stock
    dfs.append(df)

panel = pd.concat(dfs, ignore_index=True)

panel["Date"] = pd.to_datetime(panel["Date"])
panel = panel.sort_values(["Stock", "Date"])

panel.head()
```

```python
# Compute Returns
# Return is percentage change in closing price from yesterday to today
# we are using percentage rather than raw price because percentage is a better representation of price change
panel["Return"] = panel.groupby("Stock")["Close"].pct_change()
panel = panel.dropna().reset_index(drop=True)
```

```python
# Train-Test Split
# We train on past and test on future
SPLIT_DATE = "2020-01-01"

train_df = panel[panel["Date"] < SPLIT_DATE]
test_df  = panel[panel["Date"] >= SPLIT_DATE]
```

```python
# Normalization
# We normalize each stock separately because each stock has its own price scale

features = ["Open", "High", "Low", "Close", "Volume"]

for stock in panel["Stock"].unique():
    train_mask = train_df["Stock"] == stock
    test_mask  = test_df["Stock"] == stock

    # Compute mean and standard deviation using only the training data of this stock
    # This avoids using future information (no leakage)
    mu = train_df.loc[train_mask, features].mean()
    sigma = train_df.loc[train_mask, features].std()

    # Normalize training data:
    train_df.loc[train_mask, features] = (train_df.loc[train_mask, features] - mu) / sigma
    
    # Normalize test data using the SAME  and  computed from the training data:
    test_df.loc[test_mask, features] = (test_df.loc[test_mask, features] - mu) / sigma
```

```python
train_df.to_pickle("train_df.pkl")
test_df.to_pickle("test_df.pkl")
```
